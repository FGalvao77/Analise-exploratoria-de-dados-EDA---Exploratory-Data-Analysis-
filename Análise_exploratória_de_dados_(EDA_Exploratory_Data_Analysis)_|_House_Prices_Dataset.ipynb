```python
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
plt.figure(figsize=(15, 7))
plt.title('\nDistribuição dos dados da variável "SalesPrice"\n', 
          {'fontsize': 16,'fontweight': 'bold'})

sns.distplot(df['SalePrice'], 
             bins=100, hist_kws={'alpha':0.5});
```

```python
vars_num = df2.select_dtypes(exclude=['O'])
vars_num.info()
```

```python
vars_num.hist(figsize=(20,20), 
              bins=50, xlabelsize=10, ylabelsize=10);
```

```python
vars_num.corr()['SalePrice'][:-1]
```

```python
plt.figure(figsize=(22, 12))
sns.heatmap(vars_num.corr(), square=True);
```

```python
plt.figure(figsize=(22, 12))
plt.title('\nCorrelação das vars numéricas contínuas\n', 
          {'fontsize': 16,'fontweight': 'bold'})

sns.heatmap(vars_num.corr(method='pearson'), cmap='Blues', 
            vmax=1., vmin=-1., annot=True);
```

```python
vars_num_corr = vars_num.corr()['SalePrice'][:-1]

important_feats = vars_num_corr[abs(vars_num_corr) > 0.5].sort_values(ascending=False)
```

```python
print(f'Há {len(important_feats)} atributos com forte correlação com a variável alvo (SalePrice). \nSegue:')
important_feats
```

```python
for i in range(0, len(vars_num.columns), 5):
    sns.pairplot(data=vars_num, x_vars=vars_num.columns[i:i + 6], 
                 y_vars=['SalePrice']);
```

```python
import operator
```

```python
ind_feats_df = []

for i in range(0, len(vars_num.columns)-1):
    temp = vars_num[[vars_num.columns[i], 'SalePrice']]
    temp = temp[temp[vars_num.columns[i]] != 0]
    ind_feats_df.append(temp)
```

```python
all_corr = {feat.columns[0]: feat.corr()['SalePrice'][0] for feat in ind_feats_df}
all_corr
```

```python
all_corr = sorted(all_corr.items(), key=operator.itemgetter(1))

for (key, value) in all_corr:
    print(f'{key:>15}: {value:>15}')
```

```python
important_feats = [key for key, value in all_corr if abs(value) >= 0.5]

print(f'Há {len(important_feats)} atributos com forte correlação com a variável alvo (SalePrice). \nSegue:')
important_feats
```

```python
num_corr = vars_num.drop('SalePrice', axis=1).corr()

plt.figure(figsize=(20,16))
sns.heatmap(num_corr[(abs(num_corr) >= 0.5)], cmap='viridis', 
            vmax=1.0, vmin=-1.0, linewidths=.1, annot=True, 
            square=True);
```

```python
quant_feats = ['LotFrontage', 'LotArea', 'MasVnrArea', 'BsmtFinSF1', 'BsmtFinSF2', 'TotalBsmtSF', '1stFlrSF',
               '2ndFlrSF', 'LowQualFinSF', 'GrLivArea', 'BsmtFullBath', 'BsmtHalfBath', 'FullBath', 'HalfBath',
               'BedroomAbvGr', 'KitchenAbvGr', 'TotRmsAbvGrd', 'Fireplaces', 'GarageCars', 'GarageArea', 'WoodDeckSF', 
               'OpenPorchSF', 'EnclosedPorch', '3SsnPorch', 'ScreenPorch', 'PoolArea', 'MiscVal', 'SalePrice']
```

```python
quant_df = df2[quant_feats]
quant_df.head()
```

```python
feat_to_analyze = [feat for feat in quant_feats if feat in important_feats]
```

```python
feat_to_analyze.append('SalePrice')
feat_to_analyze
```

```python
fig, ax = plt.subplots(round(len(feat_to_analyze) / 3), 
                       3, figsize=(20, 15))

for i, ax in enumerate(fig.axes):
    if i < len(feat_to_analyze) - 1:
        sns.regplot(x=feat_to_analyze[i], y='SalePrice', 
                    data=df2[feat_to_analyze], ax=ax);
```

```python
cat_feats = [categ for categ in quant_feats[:-1] + 
             df2.columns.tolist() if (categ not in quant_feats) or (categ not in df2.columns.tolist())]
```

```python
df_categ = df2[cat_feats]
df_categ.head()
```

```python
df_categ.shape
```

```python
df_categ_not_num = df_categ.select_dtypes(include=['O'])
df_categ_not_num.columns
```

```python
print(f'Há {len(df_categ_not_num.columns)} atributos não numéricos. \nSegue:')
df_categ_not_num.columns.tolist()
```

```python
df_categ = pd.concat([df_categ, df2['SalePrice']], 
                     axis=1)
```

```python
print(
    df_categ['MSZoning'].unique(),
    df_categ['MSZoning'].value_counts(),
    sep='\n\n'
)
```

```python
plt.figure(figsize=(20, 9))
plt.title('\nBoxplot do atributo "MSZonging" (instâncias)\n', 
          {'fontsize': 16,'fontweight': 'bold'})

sns.boxplot(y='MSZoning', x='SalePrice', 
            data=df_categ);
```

```python
print(
    df_categ['SaleCondition'].unique(),
    df_categ['SaleCondition'].value_counts(),
    sep='\n\n'
)
```
